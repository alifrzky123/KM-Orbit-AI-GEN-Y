```python
# Main Layangan
# Requirement:
# - Gak Hujan
# - Kecepatan Angin 5 - 30 km/s

def main_layangan():
  hujan = input("lagi ujan ga? (y/n)")
  if hujan == 'y' or hujan == 'Y':
    hujan == True
    return True
  else:
    hujan == False
    kecepatan_angin = int(input('masukkan kecepatan angin'))
    if kecepatan_angin < 5 or kecepatan_angin > 30:
      print('Ayiik, gass main layangan')
    else:
      print('Skip, mending main ep ep')
    return True
  
  print(bool(hujan))
   
    
capek = main_layangan()

if capek == False:
  print("Istirahat")
else:
  print("Gass main ep ep")
```
